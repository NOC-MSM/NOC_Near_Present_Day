```python
# -- Import Python packages -- #
import glob
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

### **eORCA1 ancillary files**

```python
# -- Import eORCA1 ERA5 model domain data --#
# Define directory path to domain data:
output_dir = "/dssgfs01/scratch/otooth/npd/simulations/Domains/"
ds_mesh_mask = xr.open_dataset(output_dir + "domain_cfg.nc")
ds_subbasins = xr.open_dataset(output_dir + "subbasins.nc")
```

### **Defining Location of the RAPID-MOCHA array at 26.5N**

```python
# -- Figure -- #
plt.figure()
ds_mesh_mask.e2t.isel(t=0).where(ds_mesh_mask.mask_opensea.isel(t=0) == 1).plot()
# Add RAPID 26.5N reference line:
xs = np.arange(206, 280)
ys = np.repeat(228, len(xs))
plt.plot(xs, ys, 'k.', lw=3)

# Axes limits:
plt.xlim(175, 340)
plt.ylim(210, 235)

# Calculate average latitude:
print(ds_mesh_mask.gphiv.isel(t=0, x=slice(206, 280), y=228).mean())
```
