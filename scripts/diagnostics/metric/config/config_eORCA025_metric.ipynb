```python
# -- Import Python packages -- #
import glob
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

### **eORCA025 ancillary files**

```python
# -- Import eORCA025 ERA5 model domain data --#
# Define directory path to domain data:
output_dir = "/dssgfs01/scratch/npd/simulations/Domains/eORCA025/"
ds_mesh_mask = xr.open_dataset(output_dir + "mesh_mask.nc")
ds_subbasins = xr.open_dataset(output_dir + "subbasins.nc")
```

### **Defining Location of the RAPID-MOCHA array at 26.5N**

```python
# -- Figure -- #
plt.figure()
ds_mesh_mask.e2t.isel(time_counter=0).where(ds_mesh_mask.tmask.isel(time_counter=0, nav_lev=0) == 1).plot()
# Add RAPID 26.5N reference line:
xs = np.arange(822, 1120)
ys = np.repeat(794, len(xs))
plt.plot(xs, ys, 'k.', lw=3)

# Axes limits:
plt.xlim(700, 1200)
plt.ylim(750, 850)

# Calculate average latitude:
print(ds_mesh_mask.gphiv.isel(time_counter=0, x=slice(822, 1120), y=794).mean())
```
